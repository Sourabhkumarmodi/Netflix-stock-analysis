# NETFLIX STOCK ANALYSIS PROJECT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df=pd.read_csv("Netflix.csv")
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.Date.dtypes
```

```python
df['Month']=df['Date'].dt.month
```

```python
df
```

```python
df['Year']=df['Date'].dt.year
```

```python
df['day']=df['Date'].dt.day
```

```python
df
```

```python
df=df.set_index('Date')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=df.index,y=df['Volume'],label='Volume',color='r')
plt.title("Volume of the stock and time")
```

```python
df.plot(y=['High','Close','Open'],title='Netflic stock Analysis')
```

```python
df.groupby(df.index.day).mean().plot(y='Volume',xlabel='day')
```

```python
df.groupby(df.index.month).mean().plot(y='Volume',xlabel='month')
```

```python
df.groupby(df.index.year).mean().plot(y='Volume',xlabel='year')
```

```python
top_5_dates_with_high_stock=df.sort_values(by='High',ascending=False)
top_5_dates_with_high_stock['High'].head(5)
```

```python
Top_5_dates_with_Low_stock=df.sort_values(by='Low',ascending=True)
Top_5_dates_with_Low_stock['Low'].head(5)
```

```python
sns.lineplot(y=df['High'],x=df.index,data=df,color='g')
```

```python
sns.lineplot(x=df.index,y=df['Low'],color='r')
```
